# 🧹 Data Pre-processing

## 🔍 Checking Missing Values

Missing values can affect data quality and analytical results.

The dataset is examined to identify columns containing null values.

```python
df.isnull().sum()
```

## 🔢 Handling Missing Values in Numerical Columns

The columns `Units_Sold`, `MRP`, and `Discount_Applied` contain numerical data.

Missing values are replaced using the median because it is less sensitive to extreme values and outliers.

```python
df['Units_Sold'] = df['Units_Sold'].fillna(
    df['Units_Sold'].median()
)

df['MRP'] = df['MRP'].fillna(
    df['MRP'].median()
)

df['Discount_Applied'] = df['Discount_Applied'].fillna(
    df['Discount_Applied'].median()
)
```

## 🏷️ Handling Missing Values in Categorical Columns

The `Size` column contains missing values.

Since Size is a categorical variable, missing values are replaced using the most frequently occurring value (Mode).

```python
df['Size'] = df['Size'].fillna(
    df['Size'].mode()[0]
)
```

## 📅 Handling Missing Values in Date Column

The `Order_Date` column contains missing values.

Since date information is important for trend analysis, rows containing missing dates are removed.

```python
df['Order_Date'] = pd.to_datetime(
    df['Order_Date'],
    errors='coerce',
    format='mixed'
)
```

```python
df.dropna(
    subset=['Order_Date'],
    inplace=True
)
```

## 🔍 Checking Duplicate Records

Duplicate records can affect statistical analysis and lead to inaccurate results.

The dataset is checked for duplicate rows.

```python
print("Duplicate Records:", df.duplicated().sum())
```

## 🔄 Correcting Data Types

The `Order_Date` column is initially stored as an object data type.

To enable time-based analysis, it is converted into datetime format.

## 🆕 Creating Derived Columns

Additional features are derived from the Order_Date column to support time-based analysis.

The following columns are created:

- Year
- Month
- Quarter

```python
df['Year'] = df['Order_Date'].dt.year

df['Month'] = df['Order_Date'].dt.month_name()

df['Quarter'] = df['Order_Date'].dt.quarter
```

## 📊 Filtering and Aggregating Data

Aggregation is performed to summarize business performance and identify trends across different categories.

### Revenue by Region

The dataset is grouped by Region to calculate total revenue generated from each region.

```python
region_revenue = df.groupby(
    'Region'
)['Revenue'].sum()

print(region_revenue)
```

### Profit by Product Line

The dataset is grouped by Product_Line to determine the total profit generated by each product category.

```python
product_profit = df.groupby(
    'Product_Line'
)['Profit'].sum()

print(product_profit)
```

## ✅ Final Validation

After completing all preprocessing steps, the dataset is validated to ensure it is clean, consistent, and ready for analysis.

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Remove invalid records

df = df[df['Units_Sold'] >= 0]

df = df[df['Revenue'] >= 0]

df = df[df['Discount_Applied'] <= 1]

print("Records after validation:", len(df))
```

```python
df.to_csv('Nike_Sales_Cleaned_Final.csv', index=False)

print("Final cleaned dataset saved successfully.")
```

# Exploratory Data Analysis (EDA)

## Objective

The objective of Exploratory Data Analysis (EDA) is to understand the underlying patterns, trends, and relationships within the Nike Sales dataset. This phase uses statistical summaries, grouping techniques, and visualizations to generate meaningful business insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv("Nike_Sales_Cleaned_Final.csv")
```

```python
df.head()
```

## Dataset Overview

Before performing analysis, it is important to understand the overall structure of the dataset.

```python
print("Dataset Shape:", df.shape)

df.info()
```

```python
df.describe()
```

The cleaned dataset contains 1,594 records and 16 columns. No missing values are present, making the dataset suitable for analysis. The dataset includes information related to products, revenue, profit, discounts, sales channels, and regional performance.

## Univariate Analysis

## Objective

Univariate analysis focuses on examining a single variable at a time to understand its distribution and characteristics.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Revenue'], bins=30)

plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')

plt.show()
```

### Observation

The revenue distribution shows that most transactions generate low to moderate revenue, while a smaller number of transactions contribute significantly higher revenue values.

### Business Insight

High-revenue transactions should be analyzed to identify the products, regions, and sales channels driving business growth.

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Profit'], bins=30)

plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')

plt.show()
```

### Profit Distribution Observation

Profit values vary across transactions, indicating differences in product performance and pricing effectiveness.

### Business Insight

Products generating consistently higher profits can be prioritized for inventory management and promotional campaigns.

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Units_Sold', data=df)

plt.title('Units Sold Distribution')
plt.xlabel('Units Sold')
plt.ylabel('Count')

plt.show()
```

### Units Sold Observation

Most transactions involve a smaller number of units sold, indicating customer preference for individual purchases.

# Bivariate Analysis

## Objective

Bivariate analysis examines the relationship between two variables to identify trends and associations.

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='Revenue',
    y='Profit',
    data=df
)

plt.title('Revenue vs Profit')

plt.show()
```

### Discount vs Revenue Observation

Discounts influence customer purchasing behavior and may impact overall revenue generation.
